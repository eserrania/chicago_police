```python
%load_ext autoreload
%autoreload 2

# modules
import read_data as rd
import train_test as tt
import ml_loop as ml
import numpy as np
import full_pipeline as fp

# other libraries
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# create initial dataframes
raw_dfs = fp.RawDfs()
```

```python
# create Train-Test set list with one year outcome windows
set_list = raw_dfs.create_train_tests(np.timedelta64(1,'Y'))
```

```python
# add train and test features to each Train-Test set
for test_set in set_list:
    test_set.add_train_features()
    test_set.add_test_features()
```

```python
# Model parameters
GRID = 'all'
CLFRS_TO_RUN = ['RandomForest',
                'ExtraTrees',
                'AdaBoost',
                'DecisionTree',
                'GradientBoosting',
                'Bagging',
                'LogisticRegression']
label_lst = ['sustained_outcome', 'firearm_outcome']

# Metrics and thresholds
METRIC_DICT = {'precision': [100, 5, 20, 50, 80],
               'recall': [100, 5, 20, 50, 80],
               'f1': [5],
               'accuracy': [50]}
METRICS = 'pop'
```

```python
# handle missing values (in case fillna is not properly working, as
# fillna seems to not be working correctly sometimes)
for i in range(0, 4):
    set_list[i].reg_features.remove('unit_id_nan')
for i in range(0, 4):
    set_list[i].train = set_list[i].train.fillna(set_list[i].train.mean())
for i in range(0, 4):
    set_list[i].test = set_list[i].test.fillna(set_list[i].test.mean())
for i in range(0, 4):
    set_list[i].train = set_list[i].train.fillna(0)
for i in range(0, 4):
    set_list[i].test = set_list[i].train.fillna(0)
for i in range(0, 4):
    set_list[i].aug_features.remove('unit_id_nan')
```

```python
# feed in list of Train-Test objects and save ml loop results
res_df2 = ml.classifier_loop(set_list, GRID, CLFRS_TO_RUN, METRIC_DICT, label_lst,
                             METRICS, "big_loop_output.csv", plot=False)
```

```python
# plot results across Train-Test sets
sustained_df = res_df2[(res_df2.outcome == 'sustained_outcome') & (res_df2.type == 'augmented')].sort_values(['set', 'precision_at_5', 'model'], ascending=False)
sns.catplot(x='set', y='precision_at_5', hue='model', data=sustained_df, kind='swarm')
plt.ylabel('Precision at 5%')
plt.xlabel('Train-Test Set')
best_model = 63

def label_point(x, y, val, ax):
    a = pd.concat({'x': x, 'y': y, 'val': val}, axis=1)
    for i, point in a.iterrows():
        if point['val'] == best_model:
            ax.text(point['x']+.02, point['y'], str(point['val']))

label_point(sustained_df.set, sustained_df.precision_at_5, sustained_df.param_set, plt.gca()) 
```
